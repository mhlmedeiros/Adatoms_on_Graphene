```python
def plot_results_comparison(path):
    data_compare = np.load(path)
    trans_B = data_compare['transmission_B']
    trans_0 = data_compare['transmission_B_zero']

    fig, ax = plt.subplots(figsize=(9,6))
    ax.plot(trans_B, linewidth=2, color='blue', label=r"$B=0.5$ T")
    ax.plot(trans_0, linewidth=1, color='red', label=r"$B=0.$ T")
    ax.axhline(y=trans_0.mean(), linestyle='--', c='k')
    ax.legend(loc=0, fontsize=15)
    ax.set_xlabel('Configuration', fontsize=20)
    ax.set_ylabel(r'Conductance $[e^2/h]$', fontsize=20)
#     ax.set_ylim(-1.5, 3)
    ax.set_title("Energy = 0.01 eV", fontsize=20)
    ax.tick_params(labelsize=15)
#     plt.savefig('../images/weak_localization_1.png')
    plt.show()
    
def plot_results_difference(path):
    data_compare = np.load(path)
    trans_B = data_compare['transmission_B']
    trans_0 = data_compare['transmission_B_zero']
    difference = trans_B - trans_0
    print(difference.mean())
    fig, ax = plt.subplots(figsize=(9,6))
    ax.plot(difference, linewidth=2, color='blue', label=r"$B=0.5$ T")
#     ax.plot(trans_0, linewidth=1, color='red', label=r"$B_{flux}=0.$")
    ax.axhline(y = difference.mean(), linestyle='--', c='k')
    ax.legend(loc=0, fontsize=15)
    ax.set_xlabel('Configuration', fontsize=20)
    ax.set_ylabel(r'Conductance $[e^2/h]$', fontsize=20)
#     ax.set_yscale('log')
#     ax.set_ylim(-0.05, 0.05)
    ax.set_title("Energy = 0.01 eV", fontsize=20)
    ax.tick_params(labelsize=15)
#     plt.savefig('../images/weak_localization_1.png')
    plt.show()
    
```

## Notes:

The results presented above are promissing. Without magnetic field, the inference between the reflected wavefunction give us an conductance that is highly dependent on adatom configuration. On the other hand, when a small magnetic field is applied in the scattering region we get a higher and almost adatom-configuration independent conductance. The condunctance value jump, from the average value for systems without $\vec{B}$ and the value with magnetic field, is approximatelly $e^2/h$. Such a jump is characteristic of the weak localization phenomenon.       

```python
NN_SPACING = 0.142e-9                    ## Carbon atoms separation [m]
A_BRAVAIS = np.sqrt(3) * NN_SPACING        ## Bravais lattice spacing
PHI_0 = 2.067833848e-15                    ## [Wb] = [Tm^2]
CELL_AREA = np.sqrt(3)/2                   ## unit cell area considering a == 1
UNIT_CELL_AREA = CELL_AREA * A_BRAVAIS**2  ## m^2


def tesla_to_au(Bfield):
    """
    Convert the field unit from Tesla to PHI_0/CELL_AREA
    As the code considers A_BRAVAIS == 1 and PHI_0 == 1
    this conversion is necessary to pass the correct value
    of magnetic field to the solvers.
    """
    Bflux = field_to_flux(Bfield)  # flux of magnetic field in units of quantum of flux
    return Bflux/CELL_AREA


def field_to_flux(field):
    """
    Calculate the magnetic field flux
    in units of PHI_0.
    """
    return field * UNIT_CELL_AREA/PHI_0


def to_tesla(Bfield):
    """
    Convert the field unit from Tesla to PHI_0/CELL_AREA
    As the code considers A_BRAVAIS == 1 and PHI_0 == 1
    this conversion is necessary to pass the correct value
    of magnetic field to the solvers.
    """
    Bflux = Bfield * CELL_AREA
    return flux_to_field(Bflux)


def flux_to_field(flux):
    """
    Given a flux in units of PHI_0, returns
    the field in Tesla.
    """
    return flux * PHI_0/UNIT_CELL_AREA

```
